```python
#Question 1 pt. 4
from sklearn import linear_model
from sklearn.model_selection import train_test_split

#Subsetting dataframe
y=clean['Price']
X = clean.loc[:,['Review Scores Rating', 'Neighbourhood '] ]
#Creating Split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=100)

#Training and Test Data Sets
train = pd.concat([X_train['Review Scores Rating'],
                     pd.get_dummies(X_train['Neighbourhood '], dtype='int')], axis = 1)
test = pd.concat([X_test['Review Scores Rating'],
                    pd.get_dummies(X_test['Neighbourhood '], dtype='int')], axis = 1)

#Regression
reg3 = linear_model.LinearRegression(fit_intercept=False).fit(train, y_train)

yhat = reg3.predict(test)
rmse = np.sqrt(np.mean((y_test - yhat)**2))

print('R-squared: ', reg3.score(test, y_test))
print('RMSE: ', rmse)

fin3 = pd.DataFrame({'variable':reg3.feature_names_in_, 'coefficient': reg3.coef_})
fin3
```

```python
#Boxplot of data
sns.boxplot(data=cars)
#Unamed, make_yea, and mileage_run don't appear to show any/very little data in this original boxplot,
#indicating that transformations need to be made.

#Arcsinh Transformation
cars['Price_sinh'] = np.arcsinh(cars['Price'])
cars['Mileage_sinh'] = np.arcsinh(cars['Mileage_Run'])
#Calculation
cars['Age'] = max(cars['Make_Year']) - cars['Make_Year']
```

```python
#Question 2 pt. 3

#Randomizing rows in the dataframe:
num = cars.shape[0]
cars = cars.sample(frac=1, random_state=100)
train_size = int(0.8*num)

#Split dataset into training and test
cars_train = cars[0:train_size]
y_train = cars_train['Price_sinh']

cars_test = cars[train_size:]
y_test = cars_test['Price_sinh']
```

```python
#Question 2 pt. 4
#First Numeric Model
from sklearn import linear_model
#Choosing Variables
var_num = ['Age', 'Mileage_sinh', 'Seating_Capacity']
X_train_num = cars_train.loc[:,var_num]

#Regression
reg_num = linear_model.LinearRegression().fit(X_train_num,y_train)
X_test_num = cars_test.loc[:,var_num]

yhat = reg_num.predict(X_test_num)
rmse_num = np.sqrt(np.mean((y_test - yhat)**2))

print('R-squared: ', reg_num.score(X_test_num,y_test))
print('RMSE: ', rmse_num)
```

> The R-squared on the numeric only model is 0.47500 and the RMSE is 0.31531.

```python
#Second Categorical Model
#Choosing Variables
var_cat = ['Make','Body_Type','Color','Fuel_Type','Transmission','Transmission_Type']

#Looping through dummy df
dummies = pd.DataFrame([])
for var in var_cat:
    new_dummies = pd.get_dummies(cars.loc[:,var], drop_first=True, dtype=int)
    dummies = pd.concat([dummies, new_dummies], axis=1, ignore_index=True)
    X_train_cat = dummies.iloc[0:train_size,:]
    X_test_cat = dummies.iloc[train_size:,:]

#Regression
reg_cat = linear_model.LinearRegression().fit(X_train_cat,y_train)

yhat_cat = reg_cat.predict(X_test_cat)
rmse_cat = np.sqrt(np.mean( (y_test - yhat_cat)**2))

print('R-squared: ', reg_cat.score(X_test_cat,y_test))
print('RMSE: ', rmse_cat)
```

> The R-squared on the categorical only model is 0.63287 and the RMSE is 0.26367.

```python
#Combining numerical and categorical variables
X_train_comb = pd.concat([X_train_num, X_train_cat], axis=1, ignore_index=True)
X_test_comb = pd.concat([X_test_num, X_test_cat], axis=1, ignore_index=True)

#Combined Regression
reg_comb = linear_model.LinearRegression().fit(X_train_comb, y_train)

yhat_comb = reg_comb.predict(X_test_comb)
rmse_comb = np.sqrt( np.mean( (y_test - yhat_comb)**2 ))

print('R-squared: ', reg_comb.score(X_test_comb, y_test))
print('RMSE: ', rmse_comb)
```

> The R-squared for the combined model is 0.80288 and the RMSE is 0.19321.

> The categorical only model performs best on the test set (before the combnied model). The joint model performs the best out of all the models. Significantly better than the numerical only model, and slightly better than the cateorical only model.

```python
#Question 2 pt. 5
from sklearn.preprocessing import PolynomialFeatures

for x in np.arange(1,5):
    expand = PolynomialFeatures(degree=x,include_bias=False)
    #Expanding variables:
    Z = expand.fit_transform(X_test_num)
    vars = expand.get_feature_names_out() #Names of vars
    X_test_x = pd.DataFrame(data=Z, columns = vars) #New Dataframe
    Z = expand.fit_transform(X_train_num)
    vars = expand.get_feature_names_out() #Names of vars
    X_train_x = pd.DataFrame(data=Z, columns = vars) #New Dataframe
    X_train_x.reset_index(drop=True, inplace=True)
    X_train_cat.reset_index(drop=True, inplace=True)
    X_train_comb = pd.concat([X_train_x, X_train_cat], axis=1, ignore_index=True)

    #Reset indices
    X_test_x.reset_index(drop=True, inplace=True)
    X_test_cat.reset_index(drop=True, inplace=True)
    X_test_comb = pd.concat([X_test_x, X_test_cat], axis=1, ignore_index=True)

    #Regression
    reg_comb = linear_model.LinearRegression().fit(X_train_comb,y_train)

    yhat_comb = reg_comb.predict(X_test_comb)
    rmse_comb = np.sqrt(np.mean((y_test - yhat_comb)**2 ))
    print(x, ' Rsq: ', reg_comb.score(X_test_comb,y_test))
    print(x, 'RMSE: ', rmse_comb)
```

**Q4.** This is a math question to review the derivation of the OLS estimator (but only if you are into that kind of thing!). We are going to do it slightly differently from what we did in class, though. We will use a linear predictor and minimize the Sum of Squared Errors, just as in class. But, we are going to de-mean $X$ first, creating another variable $z_i = x_i - \bar{x}$ where
$$
\bar{x} = \dfrac{1}{N} \sum_{i=1}^N x_i,
$$
so the model is $\hat{y}_i = a + b z_i$ and the `SSE` is
$$
\text{SSE}(a,b) = \sum_{i=1}^N (y_i - a - bz_i)^2.
$$

  1. Take partial derivatives of the `SSE` with respect to $a$ and $b$. You should get

\begin{alignat*}{3}
\sum_{i=1}^N -2(y_i - a- bz_i) &=& 0 \\
\sum_{i=1}^N -2(y_i - a - bz_i)z_i &=& 0.
\end{alignat*}

  2. Solve for the solutions to the above equations. Big hint: $\bar{z} = 0$, since we subtracted the mean of $x$ from $x$ to get $z$. You should get

\begin{alignat*}{3}
a^* &=& \bar{y} \\
b^* &=& \dfrac{\sum_{i=1}^N(y_i - \bar{y})z_i}{\sum_{i=1}^N z_i^2}.
\end{alignat*}

  3. Substitute $z_i = x_i - \bar{x}$ back into the above equations. You should get
  
\begin{alignat*}{3}
a^* &=& \bar{y} \\
b^* &=& \dfrac{\sum_{i=1}^N(y_i - \bar{y})(x_i-\bar{x})}{\sum_{i=1}^N (x_i-\bar{x})^2},
\end{alignat*}

which can be written in terms of sample covariance and sample variance as:

\begin{alignat*}{3}
a^* &=& \bar{y} \\
b^* &=& \dfrac{\text{cov}(x,y)}{\text{var}(x)}.
\end{alignat*}

This is typically the preferred way of expressing the OLS coefficients.

4. When will $b^*$ be large or small, depending on the relationship between $x$ and $y$ and the amount of "noise"/variance in $x$? What does $a^*$ represent?
5. Suppose you have measurement error in $x$ which artificially inflates its variance (e.g. bad data cleaning). What happens to the $b^*$ coefficient? How will affect your ability to predict? (This phenomenon is called **attenuation**.)
